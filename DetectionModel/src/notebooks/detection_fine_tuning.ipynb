# Detection Model Fine Tuning

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[2]  
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from ultralytics import YOLO
from DetectionModel.constants.enums.yolo_variant import YoloVariant
from DetectionModel.constants.constants.detection_model import DetectionModelConstants, CHECKPOINT_DIR_PATH, resolve_dataset_yaml
from common.constants import Accelerator
```

```python
IMAGE_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif'}
DATASET_DIR = PROJECT_ROOT / 'DetectionModel' / 'datasets_clean'

def count_images(dataset_dir):
    """Count all image files in a directory, including subdirectories."""
    dataset_path = Path(dataset_dir)
    return sum(1 for f in dataset_path.rglob('*') if f.suffix.lower() in IMAGE_EXTENSIONS)

print(f"Train: {count_images(DATASET_DIR / 'train' / 'images')}")
print(f"Val:   {count_images(DATASET_DIR / 'val' / 'images')}")
print(f"Test:  {count_images(DATASET_DIR / 'test' / 'images')}")
print(f"Total: {count_images(DATASET_DIR)}")
```

```python
def train_detection_model():
    
    model = YOLO(YoloVariant.LARGE)

    # Resolve the yaml's `path:` field to the current machine's absolute path
    dataset_yaml = resolve_dataset_yaml()

    return model.train(
        data=dataset_yaml,  
        epochs=50,  
        imgsz=512,  
        batch=DetectionModelConstants.BATCH_SIZE,  
        device=Accelerator.CPU,  
        project=CHECKPOINT_DIR_PATH,  
        name=DetectionModelConstants.MODEL_FILE_NAME,  
    )
```

```python
train_detection_model()
```
