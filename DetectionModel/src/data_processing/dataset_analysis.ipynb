```python
import sys
from pathlib import Path
import json
PROJECT_ROOT = Path.cwd().parents[2]
sys.path.insert(0, str(PROJECT_ROOT))

import numpy as np
import tensorflow as tf
import pandas as pd
from collections import Counter
from constants.detection.dataset_constants import DatasetConstants
from DetectionModel.src.data_processing.data_splitter import get_patient_split
```

temp contstants defining

```python
DATASET_DIR = PROJECT_ROOT/"DetectionModel"/"datasets"

DF_DIR = DATASET_DIR/"analysis"/"nodule_analysis.csv"
PATIENTS_SPLITS_PATH = DATASET_DIR / "metadata" / "patient_splits.json"
```

loading the df and the patients splits

```python

with open(PATIENTS_SPLITS_PATH, 'r', encoding='utf-8') as file:
    patient_splits=json.load(file)

df= pd.read_csv(str(DF_DIR))
```

calculating the patient ratio of the splits

```python


split_counts = Counter(patient_splits.values())
patients_num=split_counts.total()

to_precentage = lambda split_count: (split_count/patients_num) * 100

print("".join([f'{split} : {to_precentage(count):.3f}\n' for split,count in split_counts.items()]))
```

filtering the df to include only valid nodules

```python

df = df[df["is_valid"]==True]
```

adding the split group to every patient

```python
df["split"]=df["patient_id"].map(patient_splits)
df
```

calculating the patient split ratio in the df

```python

    
splits = set(df["split"])

valid_patients_count=df.shape[0]

df_patient_splits={split:df["split"].value_counts().get(split, 0) for split in splits}



calc_ratio = lambda split_count: (split_count/valid_patients_count) * 100

print("".join([f'{split} : {calc_ratio(count):.3f}\n' for split,count in df_patient_splits.items()]))

```
